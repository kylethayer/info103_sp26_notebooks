## Loop through the list of submissions
The variable `posts` now has a list of Mastodon posts ("statuses"). So we can use a for loop to go through each post, and then use `.` to access info from each post (other pieces of information would need `[" "]` to access).

For each of the post, we will use `print` to display information about the post

```python
import re # this is a library to help us display the posts' content

for post in posts:
    print("Info for post with id: " + str(post.id))
    print("  account.username: " + str(post.account.username))
    print("  account.acct: " + str(post.account.acct))
    print("  created_at: " + str(post.created_at))

    # Note: The "content" has html tags ("<p>", "<span>", etc.)
    # print("  content: " + str(post.content))
    # we do a crude removal of the html tags using the "re" library
    # to make the post content more readable:
    print("  cleaned content: " + re.sub(r"<.*?>", "", post.content))
    
    print("  url: " + str(post.url))
    print("  favourites_count: " + str(post.favourites_count))
    print("  reblogs_count: " + str(post.reblogs_count))
    print("  quotes_count: " + str(post.quotes_count))
    print("  edited_at: " + str(post.edited_at))
    print("  tags: " + str(post.tags))
    print()
```
